```python
#### using a json of texas zip codes
texas_zip = pd.read_json('TexasZipCodes_StandardOnly.json')
texas_zip.describe()
```

Filtering the larger list of texas zip codes to only the ones near Houston and not post office boxes

```python
GreaterHouston_Zip = texas_zip.loc[texas_zip['zip'].isin(HoustonZipCodes)]
```

```python
GreaterHouston_Zip.describe()
```

```python
#### test of the .format(latT,longT) into {} {} functionality
latT = 5.2
longT = 8.4
print("q=%23actuallivingscientist&geocode={},{},4km&result_type=recent&since=2017-02-13&count=50".format(latT,longT))
```

```python
arrayOfTweetsByLocat = []
for index, row in GreaterHouston_Zip.iterrows():
    lat = row['latitude']
    long = row['longitude']
    results_forEach = api.GetSearch(raw_query="q=%23houwx&geocode={},{},10km&result_type=recent&since=2018-01-14&count=500".format(lat,long))
    arrayOfTweetsByLocat.append({"location":[lat,long],"tweets":results_forEach}) 

arrayOfTweetsByLocat
```

```python
len(arrayOfTweetsByLocat)
```

```python
arrayOfTweetsAllHou = []
for index, row in GreaterHouston_Zip.iterrows():
    lat = row['latitude']
    long = row['longitude']
    results_forEach = api.GetSearch(raw_query="q=%23houwx&geocode={},{},10km&result_type=recent&since=2018-01-14&count=500".format(lat,long))
    for status in results_forEach:
        arrayOfTweetsAllHou.append(status) 
    

len(arrayOfTweetsAllHou)
```

```python
arrayOfTweetsAllHou[2636]
```

```python
with open('houwx.json', 'w') as outfile:  
    json.dump(arrayOfTweetsAllHou, outfile)
```

### OLD CELLS FROM approach that kept the very approximate lat / long data

```python
## (status._json
arrayOfTweetsAllHouLoc = []
for index, row in GreaterHouston_Zip.iterrows():
    lat = row['latitude']
    long = row['longitude']
    results_forEach = api.GetSearch(raw_query="q=%23houwx&geocode={},{},10km&result_type=recent&since=2018-01-14&count=500".format(lat,long))
    for status in results_forEach:
        arrayOfTweetsAllHouLoc.append({"status":status,"loc":{"lat":lat,"long":long}}) 
    

len(arrayOfTweetsAllHouLoc)
```
